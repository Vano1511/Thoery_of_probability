```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])

a = (np.mean(x*y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x)**2)
b = np.mean(y) - a * np.mean(x)
a, b
```

```python
plt.scatter(x, y)
plt.plot(x, a * x + b)
plt.show
```

```python
x1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y1 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])

a1 = (np.mean(x*y1) - np.mean(x1) * np.mean(y1)) / (np.mean(x1**2) - np.mean(x1)**2)
b1 = np.mean(y1) - a1 * np.mean(x1)
a1, b1
```

```python
plt.scatter(x1, y1)
plt.plot(x1, a1 * x1 + b1)
plt.show
```

```python
x_matr = x1.reshape(len(x1), 1)
y_matr = y1.reshape(len(y1), 1)
x1_matr = np.hstack([np.ones((len(x1), 1)), x_matr])
np.dot(np.linalg.inv(np.dot(x1_matr.T, x1_matr)), x1_matr.T @ y_matr)
```

```python
model = LinearRegression()
x_r = x1.reshape(-1, 1)
# y_r = y.reshape(1, -1)
regres = model.fit(x_r, y1)
det = model.score(x_r, y1)
print("coeff of determination: ", det)
print("intercept: ", regres.intercept_)
print("slope: ", regres.coef_)
```

```python
plt.scatter(x1, y1)
plt.plot(x1, regres.intercept_ + regres.coef_ * x1)
plt.show
```
