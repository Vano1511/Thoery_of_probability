```python
# Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):
# zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. 
# Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату 
# (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). 
# Произвести расчет как с использованием intercept, так и без.

# Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

# Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов 
# должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на 
# изменение другого во время одной итерации).
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]) 
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
np.corrcoef(x, y) # проверяем линейную зависимость 0.888
```

```python
coef = (np.mean(x*y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x)**2)
coef
```

```python
plt.scatter(x, y)
plt.plot(x, coef *x)
plt.show  # без интерсепта вообще график не очень
```

```python
intercept = np.mean(y) - coef * np.mean(x)
intercept
```

```python
plt.scatter(x, y)
plt.plot(x, coef *x + intercept)
plt.show  # совсем другое дело
```

```python
#2.  Градиентный спуск (без инерсепта)

x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]) 
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
n = len(x)

def mse_(coef, x=x, y=y, n=n):
    return np.sum((coef * x - y)**2 / n)

step = 0.000001
coef = 0


for _ in range(10_000):
    coef -= step * (2 / n) * np.sum((coef * x - y) * x)
    if _ % 1_000 ==0:
        print(f"iteration = {_}, coef = {coef}, mse = {mse_(coef)}")
    
```

```python
plt.scatter(x, y)
plt.plot(x, coef *x)
plt.show  # вот что получилось
```

```python
#2.  Градиентный спуск с интерсептом

x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]) 
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
n = len(x)

def mse_(coef, intercept, x=x, y=y, n=n):
    return np.sum((coef * x + intercept - y)**2 / n)

step_coef = 0.000001
step_intercept = 0.001
coef = 0
intercept = 0.5


for _ in range(100_000):
    coef -= step_coef * (2 / n) * np.sum((coef * x + intercept - y) * x)
    intercept -= step_intercept * (2 /n) * np.sum(coef * x + intercept - y)
    if _ % 10_000 ==0:
        print(f"iteration = {_}, coef = {coef}, intercept = {intercept}, mse = {mse_(coef, intercept)}")
    
```

```python
plt.scatter(x, y)
plt.plot(x, coef *x + intercept)
plt.show  # вот что получилось с интерсептом
```
